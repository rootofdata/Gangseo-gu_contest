```python
import folium
import pandas as pd

x=[]
y=[]
for i in range(len(df['위도'])):
    if df['위도'][i]==0.0 or df['경도'][i]==0.0:
        pass
    else:
        x.append(df['위도'][i])
        y.append(df['경도'][i])
print(len(x))
```

```python
import folium
map_osm = folium.Map(lacation=[x[13],y[13]],zoom_start=10)
for i in range(len(x)):
    folium.Marker([x[i],y[i]],popup='강서구CCTV_%d'%i,icon=folium.Icon(color='red',icon='info-sign')).add_to(map_osm)
folium.CircleMarker(location=[x[13],y[13]],popup='강서구 CCTV',radius=300,color="#3186cc",fill_color="#3186cc").add_to(map_osm)

map_osm.save('서울특별시.html')
```
